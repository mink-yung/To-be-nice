```python
import pandas as pd ##pandas를 pd로 지정

data = {##딕셔너리 만들기
    "name": ["철수","영희","현서","은호","민경","서해"],
    "age": [23,21,25,25,25,10],
    "sex": ["male","female","female","male","female","female"]
}

df = pd.DataFrame(data) ##판다스 데이터프레임으로 변환 (엑셀 표처럼)
df.to_csv('students.csv' , index=False, encoding='utf-8-sig')
'''df.to_csv:CSV파일로 내보내겠다.
index=False:앞에 붙는 0,1,2 같은 번호는 빼고 저장하겠다
encoding='utf-8-sig:한글 깨뜨리지 마라'''
print(df)
print("Success")
```

📂 CSV란? (Comma-Separated Values)
이름 그대로 쉼표(,)로 구분된 값들이라는 뜻이에요.
메모장으로 열면: 데이터들이 쉼표로 다닥다닥 붙어 있어요.
엑셀로 열면: 쉼표를 기준으로 칸이 착착 나뉘어서 표처럼 보여요.

```python
with open('animals.csv','w',encoding='utf-8-sig') as f:
  f.write("종,이름,먹이,사는곳\n")
  f.write("포유류,돌고래,작은물고기,바다\n")
  f.write("파충류,뱀,달걀또는작은동물,정글\n")
  f.write("양서류,개구리,벌레,물가\n")
  f.write("포유류,말,풀또는당근,초원\n")
  f.write("양서류,도마뱀,벌레,늪\n")
  f.write("포유류,강아지,사료,인간집\n")
  f.write("포유류,밍키,해산물또는날것,은호옆\n")
'''
모드,이름,의미,특징
'w',Write (쓰기)   새로 만들기   "파일이 없으면 만들고, 있으면 내용을 싹 지우고 새로 써요. (덮어쓰기)"
'a',Append (추가)  이어서 쓰기    기존 내용 뒤에 새로운 내용을 덧붙일 때 써요.
'r',Read (읽기)    내용 보기      파일 내용을 읽기만 할 때 써요. (수정 불가)
'''
print("Success")
```

DataFrame: 여러 열을 한꺼번에 정렬하거나, 표 전체의 통계를 낼 때.

Series: 각 행의 값을 하나씩 검사하거나, 문자열을 수정(str.replace 등)할 때 편리합니다.

```python
# 단일 컬럼 선택 -> Series
name = an_df["이름"]
print(type(name))
display(name)
#여러 컬럼 선택 -> DataFrame
sub = an_df[["이름","먹이","사는곳"]]
su = an_df[["종"]]
print(type(sub))
print(type(su))
display(sub)
display(su)
```

display는 dataframe을 표로 읽어주고 print는 series처럼 글자만 긁어온다.
연산자 &,|,~ 는 각각 AND,OR,NOT을 의미. 조건들끼리는 꼭 괄호로 감싸줘야한다! 부정은 조건 앞에만 올 수 있고 접속사로 사용이 불가 하다.

```python
# 나이가 25살인 학생
df_class = st_df[st_df["age"] == 25]
print(type(df_class))
display(df_class)

# 나이가 25세 미만이고 여성인 학생
df_AND = st_df[(st_df["age"] < 25) & (st_df["sex"] =="female")]
# 나이가 25세 미만이거나 여성인 학생
df_OR = st_df[(st_df["age"] < 25) | (st_df["sex"] =="female")]
# 나이가 25세 미만이 아니고 여성인 학생
df_NOT = st_df[~(st_df["age"] < 25) & (st_df["sex"] =="female")]
display(df_AND)
display(df_OR)
display(df_NOT)
```

loc:라벨기반 인덱싱
"행 인덱스 라벨"과 "열 이름"을 기준으로 선택

```python
#0~2번 행, name과 sex 열 만
namex = st_df.loc[0:2, ["name","sex"]]
display(namex)

#특정 행 한 줄 선택
row_1 = st_df.loc[3]
display(row_1)

#조건+열 선택 조합
boys = st_df.loc[st_df["sex"] == "male", "name"]
display(boys)
```

iloc는 순수하게 "정수 위치 기반" 인덱싱
슬라이싱 팁[어디서부터:어디까지:몇개씩넘겨쓸지]


```python
#첫번째 행
first_row=st_df.iloc[0]
display(first_row)
print("----------------")
#첫 3행, 첫 2개의 열
third_two=st_df.iloc[0:3,0:2]
display(third_two)
print("----------------")
#짝수 행만 선택
even_rows = st_df.iloc[::2]
display(even_rows)
```
